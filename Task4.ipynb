```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("C:\\Users\\Krishnendu\\Downloads\\twitter_training.csv", header=None)
df.columns = ['ID', 'Entity', 'Sentiment', 'Tweet']  # Assigning column names
```

```python
# Show first few rows
print("First 5 Rows:")
print(df.head())
```

```python
# Display sentiment distribution
print("\nSentiment Distribution:")
print(df['Sentiment'].value_counts())
```

```python
# Plot sentiment distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Sentiment', data=df, order=df['Sentiment'].value_counts().index)
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Tweet Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Add a column for tweet length
df['TweetLength'] = df['Tweet'].astype(str).apply(len)
```

```python
# Boxplot of tweet length by sentiment
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sentiment', y='TweetLength', data=df)
plt.title("Tweet Length by Sentiment")
plt.ylabel("Length of Tweet")
plt.xticks(rotation=45)
plt.show()
```

```python
# Most frequent entities (brands/topics)
top_entities = df['Entity'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_entities.index, y=top_entities.values)
plt.title("Top 10 Most Mentioned Entities")
plt.xlabel("Entity")
plt.ylabel("Mentions")
plt.xticks(rotation=45)
plt.show()
```
